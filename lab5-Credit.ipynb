```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('Credit.csv')
df.head()
```

```python
target = 'Limit'
predictors = ['Income', 'Rating', 'Cards', 'Age', 'Education']
```

```python
for predictor in predictors:
    df.plot(kind='scatter', x=target, y=predictor)
```

```python
for predictor in predictors:
    plot = sns.lmplot(y=predictor, x=target, data=df)
```

```python
for predictor in predictors:
    resid = sns.residplot(target, predictor, data=df).set(xlabel=target, ylabel=predictor + ' Residuals')
    plt.figure()
```

```python
for predictor in predictors:
    linReg = LinearRegression().fit(df.filter(items=[target]), df.loc[:, predictor])
    print(predictor, '-', linReg, '\n\tcoef:', linReg.coef_, '\n\tintercept:', linReg.intercept_)
```
